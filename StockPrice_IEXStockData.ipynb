```python
from iexfinance import get_historical_data
from datetime import datetime
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
```

### Simple Example with Sine Wave

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(15*np.pi*t) 
plt.plot(s)

plt.xlabel('time (s)')
plt.ylabel('Stock Price')
plt.title('Stock Price Over Time')
plt.grid(True)
plt.show()
```

### Fourier Transform of the Sine Wave

```python
N = len(s)
T = 1.0/N

x = np.linspace(0.0, N*T, N)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

yf = np.fft.fft(s)
#plt.plot(xf, np.abs(y))
plt.plot(xf, np.abs(2.0/N * np.abs(yf[:N//2])))
plt.show()
```

```python
# Number of samplepoints
N = 600
# sample spacing
T = 1.0 / N
x = np.linspace(0.0, N*T, N)
y = 1.0*np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = np.fft.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

plt.plot(xf, np.abs(2.0/N * np.abs(yf[:N//2])))
plt.show()
```

```python
start = datetime(2019, 3, 10)
end = datetime(2019, 4, 10)

stock_data = get_historical_data("HON", start=start, end=end, output_format='json')
stock_data
```

### Plot Stock Data

```python
outArray = []
for key, value in stock_data['HON'].items():
    outArray.append(value['high'])

time_series = np.array(outArray)
plt.plot(outArray)
plt.show()
```

```python
print(len(time_series))
```

### Fourier Transform of Stock Data

```python
N = len(time_series)
T = 1.0/N
x = np.linspace(0.0, N*T, N)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
yf = np.fft.fft(np.array(time_series))
plt.plot(xf, np.abs(2.0/N * np.abs(yf[:N//2])))
plt.show()
```
