```python
def load_wordvec_model(modelName, modelFile, flagBin):
    print('Loading ' + modelName + ' model...')
    model = KeyedVectors.load_word2vec_format(model_path + modelFile, binary=flagBin)
    print('Finished loading ' + modelName + ' model...')
    return model

model_word2vec = load_wordvec_model('Word2Vec', 'GoogleNews-vectors-negative300.bin', True)
```

```python
# BUILD YOUR OWN TAXONOMY BASED ON LDA and MANUAL DATA EXPLORATION
topic_taxonomy = {
    "executive leadership manager CEO":
    {
        "C-Suite Change":    "appointed named picked joins changes quits resigns retires departs leaves moves",
        "C-Suite Compensation": "paid receives bonus payraise payout dividend"
    },
    "Strategy":
    {
        "M&A/Business Expansion":    "acquire buy merge invest purchase spend subsidize win secure contract",
        "Business Divestiture": "bankruptcy withdraw divest end exit pullout sale sell outflow writeoff split",
        "Business Partnership":  "partner team collaboration alliance deal joint venture working with",
        "Operational Outage":    "operational outage disruption halt stop cancellation delay",
        "Business Settlement":   "settlement distribute fix refund moneyback compensate reimburse",
    },
    "location branch plant store facility":
    {
        "Facility Expansion":   "open expand new",
        "Facility Relocation":  "closure move shutdown relocate"
    },
    "Industry":
    {
        "Industry":   "consumer electronics computer software phones watches ",
    },
    "stock market performance":
    {
        "Stock market performance":   "stock price increase decrease decline rise S&P shares",
    },
    "Products":
    {
        "Products":  "iPhone Apple Watch Macintosh macOS iPod iOS Keynote"
    },
    "Market Share and Competition":
    {
        "Business Competition":  "tops leads rival competitor entrant competitive advantage market share",
        "Competitors":  "Samsung Galaxy Google Pixel OnePlus Xperia Huawei Android Kindle Asus Dell Sony HP Lenovo"
    }
}
```

```python
# function checks whether the input words are present in the vocabulary for the model
def vocab_check(vectors, words):
    
    output = list()
    for word in words:
        if word in vectors.vocab:
            output.append(word.strip())
            
    return output
```

```python
# function calculates similarity between two strings using a particular word vector model
def calc_similarity(input1, input2, vectors):
    s1words = set(vocab_check(vectors, input1.split()))
    s2words = set(vocab_check(vectors, input2.split()))
    
    output = vectors.n_similarity(s1words, s2words)
    return output
```

```python
# function takes an input string, runs similarity for each item in topic_taxonomy, sorts and returns top 3 results
def classify_topics(input, vectors):
    feed_score = dict()
    for key, value in topic_taxonomy.items():
        max_value_score = dict()
        for label, keywords in value.items():
            max_value_score[label] = 0
            topic = (key + ' ' + keywords).strip()
            max_value_score[label] += float(calc_similarity(input, topic, vectors))
            
        sorted_max_score = sorted(max_value_score.items(), key=operator.itemgetter(1), reverse=True)[0]
        feed_score[sorted_max_score[0]] = sorted_max_score[1]
    return sorted(feed_score.items(), key=operator.itemgetter(1), reverse=True)[:3]
```
