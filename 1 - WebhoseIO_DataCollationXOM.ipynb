```python
# Get the next batchs of posts 
feedsxom = []
count = 0
while count < 57:
    outputxom = webhoseio.get_next()
    for feed in outputxom ['posts']:
        feedsxom.append(feed)
    count +=1
```

```python
print(len(feedsxom))
count = 0
while count < len(feedsxom):
    for feed in feedsxom:
        print(str(count+1) + " | " + str(feed['title']) + " | " + str(feed['published'][:10])) 
        count +=1 
```
