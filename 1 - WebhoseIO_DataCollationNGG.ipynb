```python
# Get the next batchs of posts 
feedsngg = []
count = 0
while count < 6:
    outputngg = webhoseio.get_next()
    for feed in outputngg ['posts']:
        feedsngg.append(feed)
    count +=1
```

```python
print(len(feedsngg))
count = 0
while count < len(feedsngg):
    for feed in feedsngg:
        print(str(count+1) + " | " + str(feed['title']) + " | " + str(feed['published'][:10])) 
        count +=1 
```
